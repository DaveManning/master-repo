```
# Load Utilities.csv, show head, and compute differences in Cost across companies.
# Then run an OLS to explain Cost and attribute differences by drivers.
import pandas as pd
from io import StringIO

# Load file
utilities_df = pd.read_csv('Utilities.csv', encoding='ascii')

# Preview head
print(utilities_df.head())

# Basic company-level cost differences
cost_stats = utilities_df[['Company','Full_Name','Cost','Fuel_Cost','Load_factor','Nuclear','Fixed_charge','Rate_of_Return','Demand_growth','Sales']].copy()

# Identify highest and lowest cost companies
top_cost = cost_stats.sort_values('Cost', ascending=False).head(5)
low_cost = cost_stats.sort_values('Cost', ascending=True).head(5)

print(top_cost)
print(low_cost)

print('Loaded data, previewed head, and computed top/bottom Cost companies.')
```

```
# Fit an OLS model to explain Cost and attribute differences; then compute per-company driver contributions
# We'll standardize predictors for comparability and compute partial contributions via beta * (x - mean_x)
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm

# Use existing utilities_df
df = utilities_df.copy()
features = ['Fixed_charge','Rate_of_Return','Load_factor','Demand_growth','Sales','Nuclear','Fuel_Cost']
X = df[features].astype(float)
y = df['Cost'].astype(float)

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=features, index=df.index)
X_const = sm.add_constant(X_scaled)
model = sm.OLS(y, X_const).fit()

# Calculate contributions (excluding intercept): beta_j * x_j_scaled
betas = model.params.drop('const')
contrib = X_scaled.mul(betas, axis=1)
contrib['Total_contrib'] = contrib.sum(axis=1)

# Summarize top and bottom cost and the largest absolute differences
df_out = df[['Company','Full_Name','Cost']].join(contrib)

# Identify extremes
top5 = df_out.sort_values('Cost', ascending=False).head(5)
bottom5 = df_out.sort_values('Cost').head(5)

# Pairwise: company with max vs min and show driver contributions for both
max_row = df_out.loc[df_out['Cost'].idxmax()]
min_row = df_out.loc[df_out['Cost'].idxmin()]

# Prepare compact tables
top5_view = top5[['Company','Full_Name','Cost','Fuel_Cost','Load_factor','Nuclear']]
bottom5_view = bottom5[['Company','Full_Name','Cost','Fuel_Cost','Load_factor','Nuclear']]

print(top5_view)
print(bottom5_view)
print(model.params)
print(model.rsquared_adj)
print('Ran OLS, computed standardized driver contributions, and summarized extremes.')
```

```
# Load data, standardize features, fit LinearRegression, report coefficients and plot standardized coefficients
# Also print R^2 and show a quick bar chart of standardized betas for interpretability
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Assume utilities_df already loaded; if not, load
try:
    df_util = utilities_df.copy()
except NameError:
    df_util = pd.read_csv('Utilities.csv', encoding='ascii')

features = ['Fixed_charge','Rate_of_Return','Load_factor','Demand_growth','Sales','Nuclear','Fuel_Cost']
X = df_util[features].astype(float)
y = df_util['Cost'].astype(float)

# Standardize X and y to interpret coefficients as effects in SD units
X_scaler = StandardScaler()
y_scaler = StandardScaler()
Xz = X_scaler.fit_transform(X)
yz = y_scaler.fit_transform(y.values.reshape(-1,1)).ravel()

lin = LinearRegression()
lin.fit(Xz, yz)

coefs = pd.Series(lin.coef_, index=features).sort_values()
rsq = lin.score(Xz, yz)

print(pd.DataFrame({'Standardized_Coefficient': coefs}))
print(rsq)

# Plot standardized coefficients
plt.figure(figsize=(7,4))
sns.barplot(x=coefs.values, y=coefs.index, orient='h', palette='viridis')
plt.title('Standardized coefficients (LinearRegression)')
plt.xlabel('Std Beta (effect on Cost SDs)')
plt.ylabel('Feature')
plt.axvline(0, color='k', linewidth=1)
plt.tight_layout()
plt.show()

print('Fitted sklearn LinearRegression with standardized variables and plotted coefficients.')
```
